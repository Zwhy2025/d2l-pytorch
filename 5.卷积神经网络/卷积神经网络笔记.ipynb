# 卷积神经网络
## 1. 卷积操作的输出形状公式为:
$$(n_h-k_h+1)\times(n_w-k_w+1)$$
其中$n_h$和$n_w$是输入数据的高度和宽度，$k_h$和$k_w$是卷积核的 高度和宽度。  
举个例子,输入数据形状：$$(14\times14)$$
卷积核形状：$$(5\times5)$$
输出形状：$$(10\times10)$$
显然得出 14-5+1=10

## 2. 填充对卷积操作的输出形状的影响：
$$(n_h-k_h+p_h+1)\times(n_w-k_w+p_w+1)\text{。}$$
  其中$p_h$和$p_w$是填充的高度和宽度。
## 3. 步幅操作对卷积操作的输出形状的影响：
$$\lfloor(n_h-k_h+p_h+s_h)/s_h\rfloor\times\lfloor(n_w-k_w+p_w+s_w)/s_w\rfloor$$
  其中$s_h$和$s_w$是步幅的高度和宽度。  
如果设置了卷积核大小与填充关系为：$p_h=k_h-1$和$p_w=k_w-1$(这个情况是最常见的，当我们希望通过卷积层不改变特征图的高度和宽度的时候就会这样填充,当然一般是为了保留边缘信息)，那么输出形状为：
 $$\lfloor(n_h+s_h-1)/s_h\rfloor\times\lfloor(n_w+s_w-1)/s_w\rfloor$$
如果输入的宽高可以被步幅整除,那么公式将会变为：
$$(n_h/s_h)\times(n_w/s_w)$$
举个例子,输入数据形状：$(14\times14)$,卷积核形状：$(5\times5)$,然后我们填充为：$(4\times4)$,这符合上面的假设.如果此时步幅为：$(2\times2)$  则输出形状为：$(7\times7)$ (ps: 这样计算的话 14/2 =7)

## 4. 多输入通道和多输出通道
- 输入通道图像形状为:$(c_i,n_h,n_w)$  
- 输入通道的卷积核形状为：$(c_i,k_h,k_w)$  
- 要使得输出通道数为：$c_0$  
- 那么卷积核的形状为：$(c_0,c_i,k_h,k_w)$  
- 输出通道图像形状为：$(c_0,m_h,m_w)$  
- (批量大小 * 通道数 * 高度 * 宽度)

## 汇聚层 
- 汇聚层的输出形状取决于汇聚窗口的大小、填充和步幅。默认情况下，汇聚窗口形状与步幅相同。导致汇聚层输出形状的宽度和高度分别为：$$(n_h/s_h)\times(n_w/s_w)$$
