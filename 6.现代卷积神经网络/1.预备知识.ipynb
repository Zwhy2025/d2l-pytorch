# 线性代数
1. **标量** -> 只有一个元素的张量
2. **向量** -> 一维数组(arange()函数)
3. **矩阵** -> 二维数组  
**一些很有意义的矩阵**
* 转置矩阵:转置矩阵的转置矩阵是原矩阵本身
* 对称矩阵:转置矩阵等于原矩阵
* 单位矩阵:对角线上的元素为1,其他元素为0
* 对角矩阵:除了对角线上的元素为1,其他元素为0
* 零矩阵:所有元素都为0
* 逆矩阵:矩阵A的逆矩阵$A^{-1}$是使得$AA^{-1}=I$的矩阵
* 正交矩阵:矩阵A的逆矩阵等于其转置矩阵$A^{-1}=A^T$
4. **张量** -> 多维数组
高阶矩阵放到一边,三维数组为图像而生!!!
* Hadamard积 两个矩阵按元素相乘
* 降维 -> sum()函数(沿着某个轴求和)
* 点积 -> dot()函数(向量对应相乘再求和)
* 矩阵-向量积 -> dot()函数(矩阵的每一行与向量相乘)
* 矩阵-矩阵乘法 -> dot()函数(矩阵的每一行与另一个矩阵的每一列相乘)
5. **范数** -> norm()函数   
非正式地说,范数是表示一个向量有多大  
1.向量的L1范数是向量元素的绝对值之和:abs()函数  
2.向量的L2范数是向量元素平方和的平方根:nurm()函数
