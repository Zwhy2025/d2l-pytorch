## 机器学习是一个训练模型的过程
训练过程通常包含如下步骤：

1. 从一个随机初始化参数的模型开始，这个模型基本没有“智能”；
2. 获取一些数据样本（例如，音频片段以及对应的是或否标签）；
3. 调整参数，使模型在这些样本中表现得更好；
4. 重复第（2）步和第（3）步，直到模型在任务中的表现令人满意。

## 机器学习的关键组件
1. 数据集  
a. 标签:需要训练的目标  
b. 样本:每个数据集you多个样本  
c. 特征:每个样本由特征向量表示，特征向量中的每个元素都是一个特征  
2. 模型  
a. 深度学习和传统机器学习的区别在于，深度学习模型更加复杂  
3. 损失函数  
a. 损失函数也叫目标函数  
b. 最常见的损失函数是均方误差（MSE）  
c. 当一个模型在训练集上的损失函数值很低，但在测试集上的损失函数值很高时，就是过拟合
4. 优化算法
a. 优化算法的作用是调整模型的参数，使损失函数的值最小化
b. 最常见的优化算法是随机梯度下降（SGD）

## 机器学习的类型
1. **监督学习**  
监督学习的目标是学习一个模型，使其能够将输入数据映射到已知的目标标签。我的理解是监督学习的模型包括"两个输入".第一个输入是类似我们后续需要预测的一部分数据,第二个是我们给数据打的标签.后续我们训练出的模型,就能够根据第一个输入,预测出第二个输入.    
*  回归问题 -> 解决有多少的问题   -> 例如预测房价 -> 平方误差损失函数  
*  分类问题 -> 解决是什么的问题   -> 例如预测猫狗 -> 交叉熵损失函数

2. **无监督学习**
无监督学习的目标是学习一个模型，使其能够对输入数据进行某种有意义的变换，这种变换可以使数据更加易于处理，也可以是数据更加符合我们的需求。我的理解是无监督学习的模型包括"一个输入".我们给模型的输入是一部分数据,模型会根据这部分数据,自己学习出一些规律,然后根据这些规律,对后续的数据进行处理.

* 主成分分析            ->

3. **与环境互动**
监督学习和无监督学习都是离线学习，即模型的训练和模型的使用是分开的。而与环境互动的学习则是在线学习，即模型的训练和模型的使用是同时进行的。  

4. **强化学习**
在强化学习问题中，智能体（agent）在一系列的时间步骤上与环境交互。在每个特定时间点，智能体从环境接收一些*观察*（observation），并且必须选择一个*动作*（action），然后通过某种机制（有时称为执行器）将其传输回环境，最后智能体从环境中获得*奖励*（reward）。此后新一轮循环开始，智能体接收后续观察，并选择后续操作，依此类推。请注意，强化学习的目标是产生一个好的*策略*（policy）。强化学习智能体选择的“动作”受策略控制，即一个从环境观察映射到行动的功能。

## 什么是神经网络
*神经网络*（neural networks）的得名源于生物灵感。类似于相互作用的神经元网络的计算电路。其核心是当今大多数网络中都可以找到的几个关键原则：
* 线性和非线性处理单元的交替，通常称为*层*（layers）；
* 使用链式规则（也称为*反向传播*（backpropagation））一次性调整网络中的全部参数。
